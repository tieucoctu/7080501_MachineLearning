Yêu cầu 2.1: Học viên đọc dữ liệu dạng excel lưu trong file excel_Data_Movies.xls theo từng sheet, sau đó nối các dataframe lại với nhau thành một DataFrame duy nhất df_movies

```python
#{HỌC VIÊN VIẾT CODE VÀO ĐÂY!}
path = 'data/excel_Data_Movies.xls'
data1 = pd.read_excel(path,
                     sheet_name=0)
data1.info()

```

```python
data2 = pd.read_excel(path,
                     sheet_name=1,
                     skiprows=3)
data2.info()
```

```python
data3 = pd.read_excel(path,
                     sheet_name = 2,
                      header=None,
                      names=['Title','Year','Genres','Language','Country','Content Rating','Duration','Aspect Ratio','Budget','Gross Earnings','Director','Actor 1','Actor 2','Actor 3','Facebook Likes - Director','Facebook Likes - Actor 1','Facebook Likes - Actor 2','Facebook Likes - Actor 3','Facebook Likes - cast Total','Facebook likes - Movie','Facenumber in posters','User Votes','Reviews by Users','Reviews by Crtiics','IMDB Score'],
                     skiprows = 3)
data3.info()
```

```python
df_movies=pd.concat([data1,data2,data3],axis=0,ignore_index=True)
df_movies.info()
```

```python
#1.Tập dữ liệu Diabetes Dataset
#{Học viên mô tả các tập dữ liệu của thư viện Sklearn vào đây!}
from sklearn.datasets import load_diabetes
import pandas as pd
diabetes = load_diabetes()
df_diabetes = pd.DataFrame(diabetes.data, columns = diabetes.feature_names)
df_diabetes.info()
```

```python
df_diabetes.head()
```

1. bộ dữ liệu chữ số viết tay.

```python
#2.Tập dữ liệu Digits Dataset
#{Học viên mô tả các tập dữ liệu của thư viện Sklearn vào đây!}
from sklearn.datasets import load_digits
import pandas as pd

digits = load_digits()
df_digits = pd.DataFrame(digits.data, columns= digits.feature_names)
df_digits.info()
```
