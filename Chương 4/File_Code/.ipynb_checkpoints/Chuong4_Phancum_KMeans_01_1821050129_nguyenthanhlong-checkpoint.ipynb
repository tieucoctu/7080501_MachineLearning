Vẽ biểu đồ hộp và biểu đồ cột để thấy số tiền chi ra của minimark với hàng tạp hóa 

```python
df['Grocery'].describe()
```

```python
#Trực quan hóa dữ liệu theo từng nhóm của minimark chi vào hàng tạp hóa:
gro_30 = df["Grocery"][(df["Grocery"] >= 0)& (df["Grocery"] <= 30000)]
gro_60 = df["Grocery"][(df["Grocery"] > 3000)& (df["Grocery"] <= 60000)]
gro_90 = df["Grocery"][(df["Grocery"] > 60000)& (df["Grocery"] <= 90000)]
gro_120 = df["Grocery"][(df["Grocery"] > 90000)& (df["Grocery"] <= 120000)]
gro_150 = df["Grocery"][(df["Grocery"] > 120000)& (df["Grocery"] <= 150000)]
aix = ["$ 0 - 30.000", "$ 30.001 - 60.000", "$ 60.001 - 90.000", "$ 90.001 - 120.000", "$ 120.001 - 150.000"]
aiy = [len(gro_30.values), len(gro_60.values), len(gro_90.values), len(gro_120.values), len(gro_150.values)]

plt.figure(figsize=(10,5))
plt.bar(aix, aiy, color='orange')
plt.title("BIỂU ĐỒ THỐNG KÊ CHI MUA TẠP HÓA")
plt.xlabel("Tổng chi mua hàng tạp hóa")
plt.ylabel("Số lượng minimark mua")
plt.grid(True)
plt.show()
```

Như vậy số lượng minimark mua hàng hóa có giá trị lớn nhất từ 30.000 đến 60.000 $.

# 2.Tiến hành phân cụm tập dữ liệu

## 2.1 Sử dụng phương pháp khửu tay tìm số cụm tối ưu

```python
#Sử dụng 3 thuộc tính Age, Income, Score để phân cụm dữ liệu
data_cluster = df.iloc[:,3:]
data_cluster.head()
```

```python
from sklearn.cluster import KMeans
wcss = []
#Thử số cụm (n_clusters) thay đổi từ 1 đến 10
for k in range(2,11):
    kmeans = KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(df.iloc[:,3:])
    wcss.append(kmeans.inertia_)

#Trực quan hóa lên biểu đồ:
plt.figure(figsize=(12,6))    
plt.grid(True)
plt.plot(range(2,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(2,11,1))
plt.ylabel("wcss")
plt.show()
```

Từ biểu đồ trên theo phương pháp khửu tay ta thấy số lượng cụm n = 5 tốt nhất

## 2.2 Tiến hành phân cụm dữ liệu theo số lượng cụm phía trên

Với k=5. Phân chia tập dữ liệu thành 5 cụm

```python
from sklearn.cluster import KMeans

#Thực hiện phân thành 5 cụm, các tham số khác sử dụng mặc định
model_km = KMeans(n_clusters=5)

#Liệt kê các tham số của mô hình
model_km.get_params()
```

```python
model_km.fit_predict(df.iloc[:,3:])

#Tạo cột Cluster để cho biết mẫu thuộc cụm nào
df["Cluster"] = model_km.labels_
df
```

```python
#Thống kê số lượng khách hàng theo mỗi cụm:
sum_cluster = df.Cluster.value_counts()
print(sum_cluster)

#Trực quan hóa dữ liệu
plt.figure(figsize=(10,4))
plt.bar(sum_cluster.index, sum_cluster.values)
plt.show()
```

```python
#Trực quan hóa phân cụm
plt.figure(figsize=(12,6))    
plt.grid(True)
plt.scatter( df["Grocery"][df.Cluster == 0], df["Milk"][df.Cluster == 0], c='blue', s=100,label='Cluster0')
plt.scatter( df["Grocery"][df.Cluster == 1], df["Milk"][df.Cluster == 1], c='red', s=100,label='Cluster1')
plt.scatter( df["Grocery"][df.Cluster == 2], df["Milk"][df.Cluster == 2], c='green', s=100,label='Cluster2')
plt.scatter( df["Grocery"][df.Cluster == 3], df["Milk"][df.Cluster == 3], c='orange', s=100,label='Cluster3')
plt.scatter( df["Grocery"][df.Cluster == 4], df["Milk"][df.Cluster == 4], c='purple', s=100,label='Cluster4')
plt.scatter(model_km.cluster_centers_[:, 0], 
           model_km.cluster_centers_[:, 1], 
           color = 'black', label = 'Centroids',s=100,marker='*')
#ax.view_init(30, 185)
plt.xlabel("Grocery")
plt.ylabel("Milk")
plt.legend()
plt.show()
```

```python
#Vị trí tâm của các cụm
np.round(model_km.cluster_centers_,1)
```

```python
#Tổng bình phương khoảng cách từ các mẫu tới trung tâm cụm gần nhất.
model_km.inertia_
```
